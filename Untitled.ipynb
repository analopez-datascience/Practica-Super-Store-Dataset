```python
import squarify
plotdata = df.market.value_counts()

plt.figure(figsize=(12, 6))
plt.axis('off')
axis = squarify.plot(sizes=plotdata.values, label=plotdata.index, alpha=.8, pad=1, text_kwargs={'fontsize': 14})

axis.set_title("Barris", fontsize=20)

plt.show()
```
